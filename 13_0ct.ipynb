````python
# prompt: make a database school  fee system

# School Fee System Database

students = []  # List to store student data

def add_student():
  """Adds a new student to the database."""
  student_id = input("Enter student ID: ")
  name = input("Enter student name: ")
  grade = input("Enter student grade: ")
  fee_amount = float(input("Enter fee amount: "))
  paid_amount = float(input("Enter amount paid: "))
  students.append({
      "student_id": student_id,
      "name": name,
      "grade": grade,
      "fee_amount": fee_amount,
      "paid_amount": paid_amount,
      "balance": fee_amount - paid_amount
  })
  print("Student added successfully!")

def view_students():
  """Displays the list of all students in the database."""
  if not students:
    print("No students found.")
  else:
    print("Student List:")
    for student in students:
      print(f"ID: {student['student_id']}, Name: {student['name']}, Grade: {student['grade']}, "
            f"Fee: {student['fee_amount']}, Paid: {student['paid_amount']}, Balance: {student['balance']}")

def search_student():
  """Searches for a student by ID."""
  student_id = input("Enter student ID to search: ")
  for student in students:
    if student["student_id"] == student_id:
      print(f"ID: {student['student_id']}, Name: {student['name']}, Grade: {student['grade']}, "
            f"Fee: {student['fee_amount']}, Paid: {student['paid_amount']}, Balance: {student['balance']}")
      return
  print("Student not found.")

def make_payment():
  """Records a payment made by a student."""
  student_id = input("Enter student ID: ")
  for student in students:
    if student["student_id"] == student_id:
      payment_amount = float(input("Enter payment amount: "))
      student["paid_amount"] += payment_amount
      student["balance"] = student["fee_amount"] - student["paid_amount"]
      print("Payment recorded successfully!")
      return
  print("Student not found.")

def main():
  """Main function to run the school fee system."""
  while True:
    print("\nSchool Fee System")
    print("1. Add Student")
    print("2. View Students")
    print("3. Search Student")
    print("4. Make Payment")
    print("5. Exit")

    choice = input("Enter your choice: ")

    if choice == "1":
      add_student()
    elif choice == "2":
      view_students()
    elif choice == "3":
      search_student()
    elif choice == "4":
      make_payment()
    elif choice == "5":
      print("Exiting...")
      break
    else:
      print("Invalid choice. Please try again.")

if __name__ == "__main__":
  main()
````
